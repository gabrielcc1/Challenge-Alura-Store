### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import math
ingresos1 = math.ceil(tienda['Precio'].sum())
ingresos2 = math.ceil(tienda2['Precio'].sum())
ingresos3 = math.ceil(tienda3['Precio'].sum())
ingresos4 = math.ceil(tienda4['Precio'].sum())

print(f'el ingreso total de la tienda 1 es ${ingresos1}')
print(f'el ingreso total de la tienda 2 es ${ingresos2}')
print(f'el ingreso total de la tienda 3 es ${ingresos3}')
print(f'el ingreso total de la tienda 4 es ${ingresos4}')

```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingresos1, ingresos2, ingresos3, ingresos4]

plt.figure(figsize=(8, 6))
plt.bar(tiendas, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'lightpink'])

plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.title('Ingresos de cada tienda')


for i, v in enumerate(ingresos):
    plt.text(i, v, str(v), ha='center', va='bottom')


plt.show()
```

# 2. Ventas por categoría

```python
ventas_categoria_tienda1 = tienda.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_categoria_tienda2 = tienda2.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_categoria_tienda3 = tienda3.groupby("Categoría del Producto")["Categoría del Producto"].count()
ventas_categoria_tienda4 = tienda4.groupby("Categoría del Producto")["Categoría del Producto"].count()


print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_tienda1)
print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_tienda2)
print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_tienda3)
print("\nVentas por categoría - Tienda 4:")
print(ventas_categoria_tienda4)
```

```python
def categoria_mas_popular(df, nombre_tienda):
    agrupado = df.groupby("Categoría del Producto")["Categoría del Producto"].count()
    categoria = agrupado.idxmax()
    unidades = agrupado.max()
    print(f"En la {nombre_tienda}: la categoria mas vendida es {categoria} con {unidades} unidades")


categoria_mas_popular(tienda, "Tienda 1")
categoria_mas_popular(tienda2, "Tienda 2")
categoria_mas_popular(tienda3, "Tienda 3")
categoria_mas_popular(tienda4, "Tienda 4")
```

```python
def categoria_mas_popular_grafico(df, nombre_tienda):
    agrupado = df.groupby("Categoría del Producto")["Categoría del Producto"].count()
    categoria = agrupado.idxmax()
    unidades = agrupado.max()

    plt.figure(figsize=(8, 6))
    plt.pie(agrupado, labels=agrupado.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Ventas por Categoría - {nombre_tienda}')
    plt.axis('equal')
    plt.show()
    print(f"En la {nombre_tienda}: la categoria mas vendida es {categoria} con {unidades} unidades")


categoria_mas_popular_grafico(tienda, "Tienda 1")
categoria_mas_popular_grafico(tienda2, "Tienda 2")
categoria_mas_popular_grafico(tienda3, "Tienda 3")
categoria_mas_popular_grafico(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python

def calificacion_promedio(df, nombre_tienda):
    promedio = df['Calificación'].mean()
    print(f"La calificación promedio de la {nombre_tienda} es {promedio}")

calificacion_promedio(tienda, "Tienda 1")
calificacion_promedio(tienda2, "Tienda 2")
calificacion_promedio(tienda3, "Tienda 3")
calificacion_promedio(tienda4, "Tienda 4")
```

```python
def mejor_calificado(df, nombre_tienda):
    df_ordenado = df.sort_values(by='Calificación', ascending=False)
    mejor_producto = df_ordenado.iloc[0]['Categoría del Producto']
    mejor_calificacion = df_ordenado.iloc[0]['Calificación']

    print(f"En la {nombre_tienda}: la categoria mejor calificada es '{mejor_producto}' con una calificación de {mejor_calificacion}")

mejor_calificado(tienda, "Tienda 1")
mejor_calificado(tienda2, "Tienda 2")
mejor_calificado(tienda3, "Tienda 3")
mejor_calificado(tienda4, "Tienda 4")
```

# 4. Productos más y menos vendidos

```python
def productos_mas_menos_vendidos(df, nombre_tienda):
    ventas_por_producto = df.groupby("Producto")["Producto"].count()
    producto_mas_vendido = ventas_por_producto.idxmax()
    unidades_mas_vendidas = ventas_por_producto.max()

    producto_menos_vendido = ventas_por_producto.idxmin()
    unidades_menos_vendidas = ventas_por_producto.min()

    print(f"En la {nombre_tienda}:")
    print(f"  Producto más vendido: {producto_mas_vendido} ({unidades_mas_vendidas} unidades)")
    print(f"  Producto menos vendido: {producto_menos_vendido} ({unidades_menos_vendidas} unidades)")

productos_mas_menos_vendidos(tienda, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
def costo_envio_promedio(df, nombre_tienda):
    promedio = round(df['Costo de envío'].mean(),2)
    print(f"El costo de envío promedio de la {nombre_tienda} es ${promedio}")

costo_envio_promedio(tienda, "Tienda 1")
costo_envio_promedio(tienda2, "Tienda 2")
costo_envio_promedio(tienda3, "Tienda 3")
costo_envio_promedio(tienda4, "Tienda 4")
```

```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [tienda['Costo de envío'].mean(), tienda2['Costo de envío'].mean(), tienda3['Costo de envío'].mean(), tienda4['Costo de envío'].mean()]

plt.figure(figsize=(10, 6))
plt.plot(tiendas, costos_envio, marker='o', linestyle='-', color='skyblue')

plt.xlabel('Tiendas')
plt.ylabel('Costo de envío promedio')
plt.title('Costo de envío promedio por tienda')
plt.grid(True)

for i, costo in enumerate(costos_envio):
    plt.text(i, costo, f"${costo:.2f}", ha='center', va='bottom')

plt.show()
```

**Informe Final de Análisis de Tiendas**

**Introducción**

El presente informe tiene como objetivo analizar el rendimiento de las cuatro tiendas en base a diversos indicadores clave de desempeño: ingresos totales, categorías de productos más y menos vendidas, productos individuales con mayor y menor volumen de ventas, calificaciones promedio de los clientes y costos de envío promedio. A partir de este análisis integral, se busca identificar cuál de las tiendas representa el menor desempeño y, por lo tanto, debería ser considerada para su posible venta.

**Desarrollo**

**1. Ingresos Totales**

* Tienda 1: \$1,150,880,400
* Tienda 2: \$1,116,343,500
* Tienda 3: \$1,098,019,600
* Tienda 4: \$1,038,375,700

La Tienda 1 lidera claramente en ingresos, mientras que la Tienda 4 presenta los ingresos más bajos del conjunto.

**2. Categorías de Productos Más y Menos Vendidas**

En todas las tiendas, la categoría más vendida es "Muebles", superando las 440 unidades en cada caso. En cuanto a las categorías menos vendidas segun el gráfico circular:

* Tienda 1: Artículos del hogar (7.2%)
* Tienda 2: Artículos del hogar (7.7%)
* Tienda 3: Artículos del hogar e Instrumentos musicales (7.5%)
* Tienda 4: Instrumentos musicales (7.2%)

Esto indica una preferencia general por productos de mobiliario, con baja rotación en artículos del hogar e instrumentos.

**3. Productos Más y Menos Vendidos (en unidades)**

* Tienda 1: Más vendido: Armario (60) / Menos vendido: Auriculares con micrófono (33)
* Tienda 2: Más vendido: Iniciando en programación (65) / Menos vendido: Juego de mesa (32)

* Tienda 4: Más vendido: Cama box (62) / Menos vendido: Guitarra eléctrica (33)

Los productos más vendidos se relacionan con el hogar, especialmente mobiliario. Los menos vendidos se concentran en tecnología o entretenimiento.

**4. Calificaciones Promedio de Clientes**

* Tienda 1: 3.98
* Tienda 2: 4.04
* Tienda 3: 4.05
* Tienda 4: 3.99

La Tienda 3 recibe las mejores calificaciones, seguida por la Tienda 2. Aunque las diferencias no son extremas, Tienda 1 está ligeramente por debajo.

**5. Costo de Envío Promedio**

* Tienda 1: \$26,018.61
* Tienda 2: \$25,216.24
* Tienda 3: \$24,805.68
* Tienda 4: \$23,459.46

La Tienda 4 tiene el menor costo de envío promedio, lo que puede representar una fortaleza operativa, aunque no se traduce en mayores ingresos o ventas.

**Conclusión**

Luego de un análisis detallado de cada tienda, se recomienda considerar la **venta de la Tienda 4**. A pesar de tener el menor costo de envío y una aceptable calificación de los clientes, sus ingresos totales son los más bajos entre las cuatro tiendas, y su producto menos vendido es de alta gama (una guitarra eléctrica), lo que sugiere una menor rotación de inventario costoso. Además, aunque comparte la categoría más vendida con las demás tiendas, no logra capitalizarla al mismo nivel que las otras en términos de ingresos.

En cambio, la Tienda 1 destaca por su alto nivel de ingresos y liderazgo en ventas, lo que indica una operación más consolidada. Las Tiendas 2 y 3 también presentan buenos indicadores en calificaciones, ventas y categorías populares.

Por lo tanto, desde una perspectiva financiera y de mercado, **la Tienda 4 es la menos competitiva y la mejor candidata para ser vendida**.
